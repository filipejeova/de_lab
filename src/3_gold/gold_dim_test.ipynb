**Criando conexão no sistema.**

```python
from google.cloud.dataproc_spark_connect import DataprocSparkSession
from google.cloud.dataproc_v1 import Session
from pyspark.sql import functions as F

# This will create a default Spark session
spark = DataprocSparkSession.builder.getOrCreate()
```

**Lendo tabela da Silver.**

```python
df = spark.read.format("bigquery").option("table", "clinic-de.2_silver.tests").load()

df.printSchema()
```

**Selecionando colunas necessarias.**

```python
df_tratado_v1 = (df.select(
    "test_code",
    "name",
    "unit",
    "low",
    "high",
    "method",
    "sample",
    "tat_hours"

    )
)
```

```python
df_tratado_v2 = df_tratado_v1.withColumnRenamed("name", "test_name") \
                 .withColumnRenamed("low", "test_low") \
                 .withColumnRenamed("high", "test_high")

```

**Salvando dataframe na Gold.**

```python
(df_tratado_v2.write
        .format("bigquery")
        .option("table", "clinic-de.3_gold.dim_test")
        .option("writeMethod", "direct")
        .mode("overwrite")
        .save()
    )

print("----------------------Tabela salva com sucesso----------------------")
```

```python
spark.stop()
```
