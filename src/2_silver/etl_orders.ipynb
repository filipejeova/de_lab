**Criando conexão no sistema.**

```python
from google.cloud.dataproc_spark_connect import DataprocSparkSession
from google.cloud.dataproc_v1 import Session
from pyspark.sql import functions as F
from pyspark.sql.functions import explode, col

# This will create a default Spark session
spark = DataprocSparkSession.builder.getOrCreate()
```

**Lendo tabela da bronze**

**Selecionando colunas do dataframe**
(A coluna "Items" foi retirada para virar um novo dataframe)

```python
query = """
SELECT
  metadata.channel AS channel,
  totals.amount   AS amount,
  totals.items    AS items,
  diagnosis,
  status,
  created_at,
  priority,
  physician_id,
  fasting,
  patient_id,
  order_id
FROM `clinic-de.1_bronze.orders`
"""

df_tratado_v1 = (
    spark.read
        .format("bigquery")
        .option("query", query)
        .load()
)
```

**Salvando dataframe na silver.**

```python
(
    df_tratado_v1
      .write
      .format("bigquery")
      .option("table", "clinic-de.2_silver.orders")
      .mode("overwrite")
      .save()
)
```

```python
spark.stop()
```
